```python
import numpy as np 
from random import shuffle

# Đọc data thô
data = []
with open("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/house.data", "r") as f:
	while True:
		line = f.readline()
		if not line:
			break
		data.append(line)

# Trả về data gồm các số thực
data1 = []
for i in data:
	new = i.split("\n")[0].split(" ")
	new_data = []
	for j in new:
		if j != '':
			new_data.append(float(j))
	data1.append(new_data)

# Trộn đều data
shuffle(data1)

# Tách data thành tập X, y
X = []
y = []
for i in data1:
	new_X = i[:13]
	new_y = i[13]

	X.append(new_X)
	y.append(new_y)	

# Tách data thành train và test theo tỷ lệ 2 : 1
train_X = X[0:len(X)//3*2]
test_X = X[len(X)//3*2:]

train_y = y[0:len(y)//3*2]
test_y = y[len(y)//3*2:]

# Save dạng nhị phân
# np.save("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/train_X.npy", train_X)
# np.save("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/test_X.npy", test_X)
# np.save("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/train_y.npy", train_y)
# np.save("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/test_y.npy", test_y)


# Save dạng thập phân
# f = open("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/train_X.txt", "w")
# for i in range (len(train_X)):
#     f.write(str(train_X[i]) + "\n")
# f.close()

# f = open("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/test_X.txt", "w")
# for i in range (len(test_X)):
#     f.write(str(test_X[i]) + "\n")
# f.close()

# f = open("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/train_y.txt", "w")
# for i in range (len(train_y)):
#     f.write(str(train_y[i]) + "\n")
# f.close()

# f = open("F:/HK4/ML/Case_Study_#2/Boston_Housing_Data/test_y.txt", "w")
# for i in range (len(test_y)):
#     f.write(str(test_y[i]) + "\n")
# f.close()
```
